```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x =[1, 2, 3, 4, 5, 6, 7]
y =[1, 2, 1, 2, 1, 2, 1]
y_error = 0.1

plt.plot(x, y)
plt.errorbar(x, y,
             yerr = y_error,
             fmt ='o')
```

```python
data = {'Category': ['A', 'B', 'C', 'D'], 'Values': [20, 34, 30, 35]}
df = pd.DataFrame(data)

plt.bar(df['Category'], df['Values'], color='skyblue')
plt.xlabel("Category")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```
